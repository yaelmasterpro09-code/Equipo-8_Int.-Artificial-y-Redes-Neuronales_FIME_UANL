```python
# importar librerias para el programa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importar dataset
Dataset=pd.read_csv("/content/robot_sensors (1).csv")
```

```python
# Mostrar encabezado
Dataset.head()
```

```python
# Mostrar tamaño de la lista sin filtros
Dimension_1 = len(Dataset)
print( "El sensor del robot capturo un total de " + str(Dimension_1) + " datos")
```

```python
# Eliminar filas vacias
Dataset_a= Dataset.dropna()
```

```python
# Mostar tamaño de la lista despues de eliminar filas vacias
Dimension_a =len(Dataset_a)
print("Total de datos sin infromacion vacia: " + str(Dimension_a))
```

```python
# Eliminar datos con velocidad negativa
Dataset_b= Dataset_a[Dataset_a['Velocidad (cm/s)']>=0]
```

```python
# Mostar tamaño de la lista despues de eliminar filas vacias y las columnas con velocidad negativa
Dimension_b =len(Dataset_b)
print("Total de datos sin velocidad negativa: " + str(Dimension_b))
```

```python
# hacer una copia de la lista con los filtros aplicados, y realizar una normalizacion a las columnas de la lista
Dataset_norm= Dataset_b.copy()
Columnas=["Distancia (cm)","Velocidad (cm/s)","Temperatura (°C)"] #definir variables para no escribir cada columna, se excluye el tiempo para facilitar la grafica
A= Dataset_b[Columnas]-Dataset_b[Columnas].min() # variable para facilitar su escritura, es el numerador de la operacion de normalizacion
B= Dataset_b[Columnas].max()-Dataset_b[Columnas].min() #denominador
Dataset_norm[Columnas]=A/B
```

```python
# Realizar la grafica de datos normalizados
plt.figure()
tiempo= Dataset_norm["Tiempo (s)"]  # Asignacion de variable para facilitar la escritura
distancia=Dataset_norm["Distancia (cm)"]  # Asignacion de variable para facilitar la escritura
velocidad=Dataset_norm["Velocidad (cm/s)"]  # Asignacion de variable para facilitar la escritura
temperatura=Dataset_norm["Temperatura (°C)"]  # Asignacion de variable para facilitar la escritura
plt.plot(tiempo,distancia, label= "distancia") #grafica de distancia
plt.plot(tiempo,velocidad, label= "velocidad") #grafica de velocidad
plt.plot(tiempo,temperatura, label= "temperatura") #grafica de temperatura
plt.xlabel("Tiempo (s)")
plt.ylabel("datos normalizados")
plt.title("grafica de datos normalizados del sensor")
plt.legend()
plt.show()
```

```python
# Grafica de prueba, con datos no normalizados
plt.figure()
tiempo1= Dataset_b["Tiempo (s)"]  # Asignacion de variable para facilitar la escritura
distancia1=Dataset_b["Distancia (cm)"]  # Asignacion de variable para facilitar la escritura
velocidad1=Dataset_b["Velocidad (cm/s)"]  # Asignacion de variable para facilitar la escritura
temperatura1=Dataset_b["Temperatura (°C)"]  # Asignacion de variable para facilitar la escritura
plt.plot(tiempo1,distancia1, label= "distancia") #grafica de distancia
plt.plot(tiempo1,velocidad1, label= "velocidad") #grafica de velocidad
plt.plot(tiempo1,temperatura1, label= "temperatura") #grafica de temperatura
plt.xlabel("Tiempo (s)")
plt.ylabel("datos no normalizados")
plt.title("grafica de datos no normalizados del sensor")
plt.legend()
plt.show()
```

si se tienen datos, no validos o sin sentido, la grafica o las funciones que se utilizaran resultara incorrecta

¿Por qué es importante la normalización en Machine Learning?

porque al normalizar la lista, es mas notorio el crecimiento de cada valor, por lo que podemos ver que tanto crece o disminuye el valor de cada columna, dentro de un mismo rango, por lo que un sistema de aprendizaje automatico podria identificar la relacion de crecimiento de cada columna con mayor exactitud

¿Cómo interpretarías los datos visualizados?

como la relacion proporcional de cambio de cada valor medido.
al observar ambas graficas es notorio que los datos de cada columna, en la grafica de datos normalizados, muestra la relacion de crecimiento de cada variable en cada instante de tiempo de forma mas equivalente , por lo que es posible observar su relacion de crecimiento de forma mas directa.
